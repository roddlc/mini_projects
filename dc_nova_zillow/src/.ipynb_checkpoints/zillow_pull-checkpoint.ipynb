# <font color = 'blue'> Section 0: Packages and CSV import

```python
# packages
import pandas as pd
import numpy as np

# display all columns
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
```

### Code below pulls static CSV file made available at [Zillow Research](https://www.zillow.com/research/data/).
Specifically, this CSV contains the Zillow Home Value Index (ZHVI) at the neighborhood level. Per the [ZHVI User Guide](https://www.zillow.com/research/zhvi-user-guide/), the ZHVI represents the "typical home value for the region".

```python
# defin var with url
url = 'https://files.zillowstatic.com/research/public_v2/zhvi/' + \
        'Neighborhood_zhvi_uc_sfrcondo_tier_0.33_0.67_sm_sa_mon.csv'

# read in neighborhood data
zhvi_raw = pd.read_csv(url)
```

```python
# first five rows of raw data
zhvi_raw.head()
```

# <font color = 'blue'> Section 1: Filter and reshape dataset

## Filter

```python
# filter to DC Metro Area
dc_area = zhvi_raw[zhvi_raw['Metro'].str.contains('Washington', na = False)]
```

```python
# show first five rows
dc_area.head()
```

## Reshape

```python
# adjust the index to facilitate reshaping the data from wide to long

# define a variable containing the column headers to set to index
init_idx_vars = ['RegionID', 'RegionName', 'RegionType', 'City',
                 'State', 'StateName', 'Metro', 'CountyName', 'SizeRank']

# set index using above columns
dc_area.set_index(init_idx_vars, inplace = True)

dc_area.head()
```

```python
# stack the table above (i.e., each year-month combination becomes a row)
dc_area = pd.DataFrame(dc_area.stack())

dc_area.head()
```

```python
# reset the index; convert the variables to lower case
dc_area = dc_area.reset_index().rename(columns = dict(zip(init_idx_vars, [v.lower() for v in init_idx_vars])))

dc_area.head()
```

```python
# rename the last two columns
dc_area.columns = [v.lower() for v in init_idx_vars] + ['month_str', 'zhvi']

dc_area.head()
```

```python
# subset to just the desired columns
idx_vars = ['regionid', 'regionname', 'city', 'state', 'countyname', 'month_str', 'zhvi']

# subset to above columns
dc_area = dc_area[idx_vars]

dc_area.head()
```

```python
# shape of table
dc_area.shape
```

# <font color = 'blue'> Section 2: Output to CSV and post-output checks

## to csv

```python
# write to csv
dc_area.to_csv("../output/dc_area_zhvi.csv", index = False)
```

## Checks
Below will check
- No `regionid`+`month_str` combination repeats
- Confirm that Virginia area displays as expected

```python
# get value counts by `regionid` and `month_str`
check1 = dc_area.groupby(['regionid', 'month_str']).size().reset_index()

# set column names
check1.columns = ['regionname', 'month_str', 'count']
```

```python
# first five rows
check1.head()
```

```python
# confirm no duplicate `regionid` + `month_str` combinations
check_2 = check1[check1['count'] > 1]

# if no duplicates, this will output an empty array
check_2['regionname'].unique()
```
